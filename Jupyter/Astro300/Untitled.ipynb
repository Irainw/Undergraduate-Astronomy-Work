```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits
from astropy.wcs import WCS
```

```python
img = "./M81_sdss_r_20230419_054440.fits"
imgData = fits.getdata(img,0)
imgHeader = fits.getheader(img,0)
wcs = WCS(imgHeader)
```

```python
fig, ax = plt.subplots(
    subplot_kw = {'projection':wcs},
    figsize = (15,5),
    constrained_layout = True
)
ax.imshow(imgData, cmap = plt.cm.magma)
```
