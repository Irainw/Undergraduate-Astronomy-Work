# Power on the Moon

<img src="images/ApolloRTG.jpg" alt="Apollo_ALSEP_RTG" width="700">

---

* The Apollo lunar mission deployed a series of experiments on the Moon.
* The experiment package was called the Apollo Lunar Surface Experiments Package [(ALSEP)](https://en.wikipedia.org/wiki/Apollo_Lunar_Surface_Experiments_Package)
* The ALSEP was powered by a radioisotope thermoelectric generator [(RTG)](https://en.wikipedia.org/wiki/Radioisotope_thermoelectric_generator)
* An RTG is basically a fist-sized slug of Pu-238 wrapped in a material that generates electric power when heated.
* Since the RTG is powered by a radioisotope, the output power decreases over time as the radioisotope decays.

---
## Read in the datafile

The data file `/Data/Apollo_RTG.csv` contains the power output of the Apollo 12 RTG as a function of time.

The data colunms are

* [Day] - Days on the Moon
* [Power] - RTG power output in Watts

Read in the datafile as a astropy `QTable`

```python
data = QTable.read('./Data/Apollo_RTG.csv', format='ascii.csv')
```

Add units to the columns

```python
data['Day'].unit = u.d
data['Power'].unit = u.W
```

```python
days = data['Day']
power = data['Power']
```

## Plot the Data

* Day vs. Power
* Fit the function with a (degree = 3) polynomial
* Plot the fit with the data
* Output size w:11in, h:8.5in
* Make the plot look nice (including clear labels)

```python
fit = np.polyfit(days.value, power.value, 3)
fitPoly = np.poly1d(fit)
data['Fit'] = fitPoly(days.value) * u.W
```

```python
fig, ax = plt.subplots(
    figsize = (11, 8.5), 
    constrained_layout = True
)

ax.set_xlabel(f"X axis ({days.unit})")
ax.set_ylabel(f"Y axis ({power.unit})")

ax.plot(days, power,
        color = "m",
        marker = "o",
        linestyle = "None",
        markersize = 10,
        label = 'Data');

ax.plot(days, data['Fit'],
        marker = "None",
        linewidth = 3,
        color = 'red',
        linestyle = '--',
        label = "Fit to Data")

ax.legend(loc=0);
```

# Fitting data to a function

* The datafile `./Data/linedata.csv` contains two columns of data [no units]

#### Read in the Data as an astropy `Qtable`

```python
data = QTable.read('./Data/linedata.csv', format='ascii.csv')
```

#### Plot the Data

* Output size w:11in, h:8.5in
* Make the plot look nice (including clear labels and a legend)

```python
waveL = data['wavelength']
flux = data['flux']

fig, ax = plt.subplots(
    figsize = (11, 8.5), 
    constrained_layout = True
)

ax.plot(waveL, flux,
        color = "m",
        marker = "x",
        linestyle = "None",
        markersize = 10,
        label = 'Data');
ax.legend(loc=0);
```

----

#### Fit a gaussian of the form:

$$ \huge f(x) = A e^{-\frac{(x - C)^2}{W}} $$

* A = amplitude of the gaussian
* C = x-value of the central peak of the gaussian
* W = width of the gaussian
* Find the values `(A,C,W)` that best fit the data

```python
def gauss (x, a, c, w):
    num = (x - c)**2
    exp = np.exp(-num / w )
    result = a * exp
    return result

```

```python
guessA = 5
guessC = 6
guessW = 3.5
initGuesses = [guessA, guessC, guessW]
fitpars, error = curve_fit(gauss,
                           waveL.value, flux.value,
                           p0 = initGuesses)
print(fitpars)
```

#### Plot the Data and the Fit on the same plot

* Output size w:11in, h:8.5in
* Make the plot look nice (including clear labels and a legend)

```python
fig, ax = plt.subplots(
    figsize = (11, 8.5), 
    constrained_layout = True
)

ax.plot(waveL, flux,
        color = "m",
        marker = "o",
        linestyle = "None",
        markersize = 10,
        label = 'Data');

ax.plot(waveL, gauss(waveL.value, *fitpars),
        color = "r",
        marker = "None",
        linestyle = "-",
        markersize = 10,
        label = 'Fit to Data');

ax.legend(loc=0);
```

# Stellar Spectra

#### The file `./Data/StarSpectra_Data.csv` is a spectra of a main sequence star

* Col 1 - Wavelength `[angstroms]`
* Col 2 - Normalized Flux `[no units]`

#### Read in the Data as an astropy `Qtable`

```python
data = QTable.read('./Data/StarSpectra_Data.csv', format='ascii.csv')
```

#### Add units to the `Wavelength` column

```python
data['wavelength'].unit = u.Angstrom
```

#### Plot the Data

* Output size w:11in, h:8.5in
* Make the plot look nice (including clear labels and a legend)

```python
waveL = data['wavelength']
flux = data['flux']

fig, ax = plt.subplots(
    figsize = (11, 8.5), 
    constrained_layout = True
)

ax.plot(waveL, flux,
        color = "m",
        marker = ".",
        linestyle = "None",
        markersize = 5,
        label = 'Data');

ax.set_xlabel(f"X axis ({waveL.unit})")
ax.set_ylabel(f"Y axis ({flux.unit})")

ax.legend(loc=0);
```

### Find the wavelength where the Flux is at a maximum

```python
bigFlux = data[(flux.max() == data['flux'])]
bigFluxWL = bigFlux['wavelength'][0]
# my_data3[(my_y3.max() == my_data3['flux'])]['wavelength'][0]
```

### Use [Wien's law](https://en.wikipedia.org/wiki/Wien%27s_displacement_law) to determine the temperature of the Star

* Use the Astropy units and constants - do not hardcode

```python
def starMAN (wavelength):
    result = const.b_wien /  wavelength
    return result.decompose()
```

```python
bigWavelength = bigFluxWL
starTemp = starMAN(bigWavelength)
print(f'The temperature of the star is {starTemp:.2f}')
```

### [Plank's Law](https://en.wikipedia.org/wiki/Planck%27s_law)

* [Plank's Law](https://en.wikipedia.org/wiki/Planck%27s_law) describes the spectra emitted by a blackbody at a temperature T
* You will want to look at the $\large \lambda$ version
* Hint: all of the units should cancel in the `exp()` part of the expression  
* Write a function to calculate the blackbody flux, at the above temperature, for all of your data_wavelength points
* Use the Astropy units and constants - do not hardcode
* Scale the blackbody flux to `[0->1]`
* Add a column to the table: `Blackbody`

```python
# Write a function
def bbFLUX(wavelength, temp):
    exp =  np.exp((const.h * const.c)/(wavelength * const.k_B * temp))
    first = 1/(exp -1)
    second = (2* const.h * const.c **2)/(wavelength**5)
    result = first * second
    return result.decompose()
```

```python
# Apply the function
wavelength = data['wavelength']
bb = bbFLUX(wavelength, starTemp)
```

```python
# Normalize and add column
def normalized(bb):
    result = (bb - bb.min()) / (bb.max() - bb.min())
    return result
bbNorm = normalized(bb)
data['Blackbody'] = bbNorm
```

### Plot the Data and the Blackbody fit on the same plot

* **Your blackbody fit should match the data pretty well.**
* Output size w:11in, h:8.5in
* Make the plot look nice (including clear labels and a legend)

```python
bbodies = data['Blackbody']

fig, ax = plt.subplots(
    figsize = (11, 8.5), 
    constrained_layout = True
)

ax.plot(waveL, flux,
        color = "m",
        marker = ".",
        linestyle = "None",
        markersize = 5,
        label = 'Data');
ax.plot(waveL, bbodies,
        color = "r",
        marker = "None",
        linestyle = "-",
        markersize = 5,
        label = 'Blackbody Fit');

ax.set_xlabel(f"X axis ({waveL.unit})")
ax.set_ylabel(f"Y axis Flux ({flux.unit})")

ax.legend(loc=0);
```
