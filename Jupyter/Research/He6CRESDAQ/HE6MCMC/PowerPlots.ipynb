```python
def findClosest(freqNoise, TE11Noise, dfNoise, dfClean, freqCol, TE11Col):
    #Establish a minimum difference between freq and TE11 power and then set the row with the smallest difference as the closest
    minimumDiff = dfClean[freqCol][0] + dfClean[TE11Col][0]
    closestColumnIndex = 0
    combinedDiff = 0
    for i in range(len(dfNoise[freqCol])):
        freqDiff = freqNoise[0] - dfClean[freqCol][i] #difference between the freqNoise and the actual frequency at index i
        TE11Diff = TE11Noise[0] - dfClean[TE11Col][i] #difference between the TE11Noise and the actual TE11Power at index i
        combinedDiff = freqDiff + TE11Diff
        if abs(minimumDiff) > abs(combinedDiff):
            minimumDiff = combinedDiff
            closestColumnIndex = i
            #If the minimum difference is greater than the new combined difference set the minimum difference to be the new combined difference
            #Set closest equal to the current index
    return dfClean.iloc[closestColumnIndex] #should be a row or an index of a row
```

```python
noiseTE11 = []
noiseFreq = []

mu, sigma = 0, 0.05e-15

for freq in frequency:
    # Generate a random percentage between 1 and 10
    percent = random.uniform(0.01, 0.1)
    noise = freq * percent #Noise is 1-10% of frequency
    
    newFreq = freq + noise #Add the noise to the frequency
    
    noiseFreq.append(newFreq)

for power in te11Power:
    samples = makeGaussError(1, mu, sigma)
    sqrtPower = np.sqrt(power)
    noise = (sqrtPower + samples) ** 2
    
    newPower = power + noise
    noiseTE11.append(newPower)
```

```python
from scipy.stats import gaussian_kde

# Sample data (x, y coordinates of points on the circle)
# Angle values from 0 to 2*pi
r = eid07['rho'][0:10000]
theta = np.linspace(0,(2*np.pi), len(r))
x = []
y = []

for radius in r:
    x.concatenate

# Concatenate the x and y coordinates to represent 2D points
data = np.vstack([x, y])

# Perform kernel density estimation
kde = gaussian_kde(data)

# Generate a grid of points on the circle
grid_points = np.vstack([x, y])

# Calculate the density estimate for each point on the grid
density_estimate = kde(grid_points)

# Plot the density estimate in polar coordinates
ax = plt.subplot(111, polar=True)
ax.plot(theta, density_estimate)
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_rlabel_position(0)  # Move radial labels to the inside
plt.title('Probability Density Distribution on the Circle (Polar Plot)')
plt.show()
```
