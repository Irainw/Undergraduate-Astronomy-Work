# Creating the model using MCMC Sampling via the emcee package

Setting up the walkers

```python
def logPrior(theta):
    Om, Od, H0 = theta
    if (0<Om<1 and 0<Od<1 and 60<H0<120):
        return 0.0 
    else:
        return -np.inf

def logLikelihood(theta, data):
    Om, Od, H0 = theta
    if (0<Om<1 and 0<Od<1 and 60<H0<120):
        #print ("here")
        cosmodel = cosmo.LambdaCDM(H0=H0,Om0=Om,Ode0=Od)
        observed = data['supernova_distance'].data
        expected = cosmodel.luminosity_distance(data['#Redshift'])
        model = expected.value
        sigma = data['supernova_distance_error'].data
        return -(np.sum((observed-model)/sigma)**2) #Return -χ^2 emcee takes exp(-χ^2)
    else:
        return -np.inf

def logPosterior(param, data):
    return logPrior(param) + logLikelihood(param, data) #emcee needs a function that returns the natural log of the posterior
```

Setting up the walkers

```python
#Initial guesses of Using emcee
ndim = 3 #Number of parameters in the model
nwalkers = 20 #Number of walkers
nBurn = 1000 #Burn off 1000 periods to let the chain stabilize
nSteps = 9000 #Number of MCMC steps
theta = [0.2, 0.8, 85.4] #inital guess data Ωm, Ωd, H0

initial = np.array(theta + 0.01 * np.random.rand(nwalkers, ndim)) #Create multiple starting positions with some random offset from theta.
sampler = emcee.EnsembleSampler(nwalkers, ndim, logPosterior, args=[snData]) #set up the sampler
sampler.run_mcmc(initial, nSteps, progress = True) #run the sampler starting at the initial position taking nSteps and showing progress towars completion

print('Done')
```

Creating a trace of where the walkers went

```python
#Plot the path of the various walkers, current problem is that they do not converge
fig, axes = plt.subplots(3, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()
labels = ["Ωm", "Ωd", "H0"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    #ax.set_ylim(-0.01,0.1)
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```

```python
#Creating a function to model the luminosity distance
def modeling(theta, redshift):
    Om, Od, H = theta
    cosmodel = cosmo.LambdaCDM(H0 = H, Om0 = Om, Ode0 = Od)
    lumDist = cosmodel.luminosity_distance(redshift)
    return lumDist
# Causes errors because cosmo needs float but PyMC3 passes in a different variable type
```
