```python
import matplotlib.pyplot as plt

import numpy as np
import astropy as ap
import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.time import Time
import pytz

from astroplan.plots import plot_airmass
from astroplan import Observer, FixedTarget, time_grid_from_range
from astroplan.plots import plot_sky
from astroplan import observability_table
from astroplan import AltitudeConstraint, MoonSeparationConstraint, SunSeparationConstraint

import warnings
warnings.filterwarnings("ignore")

from astroplan import download_IERS_A
download_IERS_A()
```

```python
APO = Observer.at_site('APO')
obsDate1 = Time("2023-5-20")
obsDate2 = Time("2023-5-21")
obsDate3 = Time("2023-5-22")
obsDate4 = Time("2023-5-23")
obsDate5 = Time("2023-5-24")

obsDateJune = Time("2023-6-8")
APONoonJune = APO.noon(obsDateJune, which = "nearest")
APOObsTimeJune = APO.tonight(APONoonJune, horizon = -18*u.deg)

APONoon20 = APO.noon(obsDate1, which = "nearest")
APONoon21 = APO.noon(obsDate2, which = "nearest")
APONoon22 = APO.noon(obsDate3, which = "nearest")
APONoon23 = APO.noon(obsDate4, which = "nearest")
APONoon24 = APO.noon(obsDate5, which = "nearest")

APOObsTime20 = APO.tonight(APONoon20, horizon = -18 * u.deg)
APOObsTime21 = APO.tonight(APONoon21, horizon = -18 * u.deg)
APOObsTime22 = APO.tonight(APONoon22, horizon = -18 * u.deg)
APOObsTime23 = APO.tonight(APONoon23, horizon = -18 * u.deg)
APOObsTime24 = APO.tonight(APONoon24, horizon = -18 * u.deg)

print(f'Observing time starts at {APOObsTime20[0].iso} UTC and ends at {APOObsTime20[1].iso} UTC')
print(f'Observing time starts at {APOObsTime21[0].iso} UTC and ends at {APOObsTime21[1].iso} UTC')
print(f'Observing time starts at {APOObsTime22[0].iso} UTC and ends at {APOObsTime22[1].iso} UTC')
print(f'Observing time starts at {APOObsTime23[0].iso} UTC and ends at {APOObsTime23[1].iso} UTC')
print(f'Observing time starts at {APOObsTime24[0].iso} UTC and ends at {APOObsTime24[1].iso} UTC')
```

```python
BH261_coord = SkyCoord(ra=168.5275 * u.deg, dec=-28.635*u.deg)
BH261 = FixedTarget(coord=BH261_coord, name="BH 261")
FSR1775_coord = SkyCoord(ra=269.0329167 * u.deg, dec = -36.565 * u.deg)
FSR1775 = FixedTarget(coord = FSR1775_coord, name = "FSR 1775")
Gran4_coord = SkyCoord(ra= 278.113 * u.deg, dec= -23.114*u.deg)
Gran4 = FixedTarget(Gran4_coord, name = "Gran 4")


M3 = FixedTarget.from_name('M3')
M53 = FixedTarget.from_name('M53')
ngc5466 = FixedTarget.from_name('ngc5466')
ngc188 = FixedTarget.from_name('ngc188')
ngc6441 = FixedTarget.from_name('ngc6441')
ngc6569 = FixedTarget.from_name('ngc6569')
ngc6380 = FixedTarget.from_name('ngc6380')
Patchick99 = FixedTarget.from_name('Patchick 99')
Ter12 = FixedTarget.from_name('Ter 12')
ESO45609 = FixedTarget.from_name('ESO 456-09')
ESO39312 = FixedTarget.from_name('ESO 393-12')
Djorg1 = FixedTarget.from_name('Djorg 1')
ngc6453 = FixedTarget.from_name('ngc6453')
ngc6522 =FixedTarget.from_name('ngc6522')
ngc6528 = FixedTarget.from_name('ngc6528')
ngc6558 = FixedTarget.from_name('ngc6558')
Ter9 = FixedTarget.from_name('Ter9')
Djorg2 = FixedTarget.from_name('Djorg2')
Ter10 = FixedTarget.from_name('Ter10')
ngc6540 = FixedTarget.from_name('ngc6540')
ngc6544 = FixedTarget.from_name('ngc6544')
ngc6553 = FixedTarget.from_name('ngc6553')

fullTargetList = [ngc6441, ngc6569, ngc6380, Patchick99, Ter12, ESO45609, ESO39312, Djorg1, ngc6453, Gran4,ngc6522, ngc6528, ngc6558, Ter9, Djorg2, Ter10, ngc6540, ngc6544, ngc6553, BH261, FSR1775]
targetList1 = [ngc6441, ngc6569, ngc6380, Patchick99, Ter12, ESO45609, ESO39312, Djorg1, ngc6453, Gran4]
targetList2 = [ngc6522, ngc6528, ngc6558, Ter9, Djorg2, Ter10, ngc6540, ngc6544, ngc6553, BH261, FSR1775]
```

```python
#Airmass Plots
timeGrid = time_grid_from_range(APOObsTime21, time_resolution = 0.01 * u.h)

fig, ax = plt.subplots(
    figsize = (15, 8),
    constrained_layout = True
)

for obsObj in targetList1:
    ax = plot_airmass(obsObj, APO, timeGrid)
    
ax.legend(loc = 0, shadow = True);
```
