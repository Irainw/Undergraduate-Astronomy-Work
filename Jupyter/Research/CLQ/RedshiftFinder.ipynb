```python
import pandas as pd
```

```python
data = pd.DataFrame()
elements = ['Lya', 'NV', 'SiIV/OIV', 'CIV', 'CIII', 'MgII', 'OII', 'Hy', 'HB', 'OIII', 'OIII', 'Ha']
restWavelength = [1216, 1240, 1400, 1549, 1909, 2800, 3727, 4340, 4861, 4959, 5007, 6563]
comments = ['Broad','Broad','Broad','Broad', 'Broad', 'Broad', 'Narrow', 'Broad, weak', 'Broad', 'Two lines near each other, often Narrow',
            'Two lines near each other, often Narrow', 'Broad']


data['Elements'] = elements
data['Rest Wavelength'] = restWavelength
data['Comments'] = comments
```

```python
def redFinder(df, z):
    wL = df['Rest Wavelength']
    redshiftWL = []
    for elements in wL:
        redshiftWL.append(elements * z + elements)
    df['Redshifted Wavelengths'] = redshiftWL
    return df
```

```python
redshifts = redFinder(data, 1.3849)
redshifts
```
