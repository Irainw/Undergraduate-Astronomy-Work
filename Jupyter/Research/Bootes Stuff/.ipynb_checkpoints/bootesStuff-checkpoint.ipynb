```python
import matplotlib.pyplot as plt

import numpy as np
import astropy as ap
import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.time import Time
import pytz

from astroplan.plots import plot_airmass
from astroplan import Observer, FixedTarget, time_grid_from_range
from astroplan.plots import plot_sky
from astroplan import observability_table
from astroplan import AltitudeConstraint, MoonSeparationConstraint, SunSeparationConstraint

import warnings
warnings.filterwarnings("ignore")

from astroplan import download_IERS_A
download_IERS_A()
```

```python
MRO = Observer.at_site('MRO')
```

```python
obsDate1 = Time("2023-6-16")
obsDate2 = Time("2023-6-17")
obsDate3 = Time("2023-6-18")
obsDate4 = Time("2023-6-19")

dates = [obsDate1, obsDate2, obsDate3, obsDate4]
obsTimes = []
for date in dates:
    noon = MRO.noon(date, which = "nearest")
    obsTime = MRO.tonight(noon, horizon = -18 * u.deg)
    obsTimes.append(obsTime)
```

```python
tvBootes = FixedTarget.from_name('tv bootes')
tvBootess = [tvBootes, tvBootes]
```

```python
timeGrids = []
for obsTime in obsTimes:
    timeGrid = time_grid_from_range(obsTime, time_resolution = 0.01 *u.h) # I hate python function naming convention
    timeGrids.append(timeGrid)
```

```python
fig, ax = plt.subplots(
    figsize = (15,8),
    constrained_layout = True
)
ax = plot_airmass(tvBootes, MRO, timeGrids[0])
ax.legend(loc = 0, shadow = True);
```

```python
fig, ax = plt.subplots(
    figsize = (15,8),
    constrained_layout = True
)
ax = plot_airmass(tvBootes, MRO, timeGrids[1])
ax.legend(loc = 0, shadow = True);
```

```python
fig, ax = plt.subplots(
    figsize = (15,8),
    constrained_layout = True
)
ax = plot_airmass(tvBootes, MRO, timeGrids[2])
ax.legend(loc = 0, shadow = True);
```

```python
fig, ax = plt.subplots(
    figsize = (15,8),
    constrained_layout = True
)
ax = plot_airmass(tvBootes, MRO, timeGrids[3])
ax.legend(loc = 0, shadow = True);
```

```python
obsRanges = []
for i in range(len(obsTimes)):
    obsRange = time_grid_from_range(obsTimes[i], time_resolution = 1 * u.h) # I hate python function naming convention
    obsRanges.append(obsRange)
```

```python
for i in range(len(obsRanges)):
    plot_sky(tvBootes, MRO, obsRanges[i])
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5));
```

```python
constraints = [SunSeparationConstraint(6* u.hourangle), MoonSeparationConstraint(60*u.deg), AltitudeConstraint(min=27*u.deg, max = 90* u.deg)]
for obsTime in obsTimes:
    obsTable = observability_table(constraints, MRO, tvBootess, time_range = obsTimes[1])
    obsTable.sort('fraction of time observable', reverse = True)    
    print(obsTable)
```
