# Question 7

From Professor Davenport's lectures I found that P(r) = 2/3Ï€Ï^2G[R^2 - r^2].

```python
import astropy.units as u
from astropy.visualization import quantity_support
quantity_support()  

grav = 6.67430e-8 * ((u.dyn * (u.cm**2))/(u.g**2))

rho = 1 * (u.g/(u.cm**3))

R = 1.004465 * u.solRad
Rcm = R.to(u.cm)
sqrR = Rcm**2
r = profile147['radius'] * u.solRad
radiusDiff = []

pressureHydro = []

for i in range(len(r)):
    rcm = r[i].to(u.cm)
    sqrr = rcm**2
    diff = sqrR - sqrr
    pressure = (2/3 * np.pi * grav * rho **2 * (diff))
    pressureHydro.append(pressure.value)
     
profileR = []
for i in range(len(profile147['radius'])):
    radius = profile147['radius'][i] * u.solRad
    radiuscm = radius.to(u.cm)
    profileR.append(radiuscm.value)

fig, ax = plt.subplot_mosaic(
    '''AB''',
    figsize = (12,8),
    constrained_layout = True
)
ax['A'].plot(profile147['radius'], profile147['logP'], label = 'Radius vs Pressure MESA')
ax['A'].set_xlabel('Radius(Rsun)')
ax['A'].set_ylabel('Log(P)')

ax['B'].plot(profileR, np.log10(pressureHydro), label = 'Radius vs Pressure via Hydrostatic Equilibrium')
ax['B'].set_xlabel('Radius(cm)')
ax['B'].set_ylabel('Log(P)');
```

7) A key difference between the calculations I did and the calculations from MESA are that I assumed a constant density and MESA did not. Additionally I converted my units to cgs and didn't convert them back while MESA had theirs in solar radii.
