```python
temp = historyData1M['log_Teff']
lum = historyData1M['log_L']
fig, ax = plt.subplots(
    figsize = (12, 8),
    constrained_layout = True
)
ax.plot(temp, lum, color = 'm')
ax.set_xlabel('log(Teff)')
ax.set_ylabel('log(Luminosity)')

ax.annotate('End of ZAMS transition to MS', xy=(3.748698798, -0.15490196), xycoords='data',
            xytext=(3.73, -0.4), textcoords='data',
            arrowprops=dict(facecolor='black'))
ax.plot(3.748698798, -0.15490196,color = 'black',marker = 'o')

ax.annotate('End of MS transition to Red Giant', xy=(3.717854332, 0.3263358609), xycoords='data',
            xytext=(3.75, 1), textcoords='data',
            arrowprops=dict(facecolor='red'))
ax.plot(3.717854332, 0.3263358609, color = 'red',marker = 'o')

ax.annotate('Transitions to horizontal branch', xy=(3.644, 1.896), xycoords='data',
            xytext=(3.76, 2), textcoords='data',
            arrowprops=dict(facecolor='orange'))
ax.plot(3.644, 1.896, color = 'orange',marker = 'o')

ax.annotate('Star ejects envelope', xy=(3.589, 3.018), xycoords='data',
            xytext=(3.67, 3), textcoords='data',
            arrowprops=dict(facecolor='green'))
ax.plot(3.589, 3.018, color = 'green',marker = 'o')

ax.annotate('Present Day Sun', xy=(3.761, 0), xycoords='data',
            xytext=(3.7, 0), textcoords='data',
            arrowprops=dict(facecolor='white'))
ax.plot(3.761, 0, color = 'black',marker = 'o')

ax.invert_xaxis();
```

The phase that we are missing in this diagram is the transition from the AGB phase where the stars envelope is ejected to the planetary nebula stage/white dwarf stage. 

# Question 2

## What happens when the surface temperature drops, but the luminosity increases?

In snapshot 290 the H in the core has been exhausted. At this point the core has stopped fusing hydrogen in and so there is no longer any outward radiation pressure which causes the helium core to contract. This heats it up enough that the hydrogen shell around the core can begin fusing which causes the radius of the star to expand which cools it down.

# Question 3

## The star continues to brighten, why?

The short answer is that the star is emitting light over a larger surface area and so will appear brighter. To be more precise, using the known luminosity equation that relates radius and temperature the radius increases by 100 times, but the temperature drops by only around 60%.

# Question 4

## What happens in the interior of the star after it peaks in luminosity and heats up?

At this point the core of the star has heated up enough to cause the Helium in the core to have enough energy to overcome strong repulsion and in a flash begin to fuse.

## Plot radius vs time

```python
time = historyData1M['star_age']
radius = historyData1M['log_R']
logTime = np.log10(time[567] - time)
```

```python
fig, ax = plt.subplots(
    figsize = (12,8),
    constrained_layout = True
)

ax.annotate('ZAMS', xy=(10.1, 0.715), xycoords='data',
            xytext=(10.1, 1), textcoords='data',
            arrowprops=dict(facecolor='white'))
ax.plot(10.1, 0.715, color = 'black',marker = 'o')

ax.annotate('Start of Main Sequence', xy=(10.1, -0.05), xycoords='data',
            xytext=(8, -0.05), textcoords='data',
            arrowprops=dict(facecolor='white'))
ax.plot(10.1, -0.05, color = 'black',marker = 'o')

ax.annotate('End of Main Sequence', xy=(9.9, 0.005), xycoords='data',
            xytext=(9.9, 0.5), textcoords='data',
            arrowprops=dict(facecolor='white'))
ax.plot(9.9, 0.005, color = 'black',marker = 'o')

ax.annotate('End of Helium Flash', xy=(8.1, 0.875), xycoords='data',
            xytext=(7, 0.75), textcoords='data',
            arrowprops=dict(facecolor='white'))
ax.plot(8.1, 0.875, color = 'black',marker = 'o')

ax.annotate('Envelope Ejection', xy=(3.6, 1.86), xycoords='data',
            xytext=(5, 1.5), textcoords='data',
            arrowprops=dict(facecolor='white'))
ax.plot(3.6, 1.86, color = 'black',marker = 'o')

ax.invert_xaxis()
ax.plot(logTime, radius);
```

In some cases it is pretty obvious where the star is happening in the core or in a shell, I can make some educated guesses from the graph, but no it is not possible for me to know exactly where the star is burning in a shell or in a core from this diagram. I can recognize the location of the helium flash as it is the point where the star rapidly expands and then suddenly contracts. 

The star will expand from the end of the main sequence until the helium flash as during this time helium is being deposited on the core from the hydrogen shell's fusion which causes the whole star to expand, once there is enough heat in the helium core to allow helium to fuse it will suddenly collapse and allow fusion to take place.
