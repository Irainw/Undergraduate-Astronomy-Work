# Question 1

```python
import matplotlib.pyplot as plt
from astropy.table import QTable
import astropy.units as u
import numpy as np
```

```python
sunData = QTable.read('./sun-atmosphere-kamp.csv', format = 'ascii.csv')
vegaData = QTable.read('./vega-atmosphere-kampV2.csv', format = 'ascii.csv')
```

```python
def thetaTemp(theta):
    temp = 5040/theta
    return temp
```

```python
sunTemp = thetaTemp(sunData['theta'])
vegaTemp = thetaTemp(vegaData['theta'])
sunData['temp'] = sunTemp
vegaData['temp'] = vegaTemp
```

```python
U1 = 2
U2 = 1
xion = 13.595
```

```python
def sahaEqn(xion, temp, theta, logPE):
    result = (5/2)*np.log10(temp) - xion * theta - logPE - 0.4772
    return result
```

```python
sunData['n1n2'] = sahaEqn(xion, sunData['temp'], sunData['theta'], sunData['logPE'])
```

```python
vegaData['n1n2'] = sahaEqn(xion, vegaData['temp'], vegaData['theta'], vegaData['logPE'])
```

```python
sunX = sunData['temp']
sunY = sunData['n1n2']
vegaX = vegaData['temp']
vegaY =vegaData['n1n2']


fig, ax = plt.subplot_mosaic(
    '''
    AB
    ''',
    figsize = (15, 8),
    constrained_layout = True
)

plt.sca(ax['A'])
plt.scatter(sunX, sunY)
ax['A'].set_xlabel('Sun-like Temperature (K)')
ax['A'].set_ylabel('log(NII/NI) of H')
ax['A'].set_title('Ratio of Hydrogen Ions vs Temperature in a Sun-like Star')

plt.sca(ax['B'])
ax['B'].set_xlabel('Vega-like Temperature (K)')
ax['B'].set_ylabel('log(NII/NI) of H')
plt.scatter(vegaX, vegaY)
ax['B'].set_title('Ratio of Hydrogen Ions vs Temperature in a Vega-like Star');
```

# Question 2

```python
PE = 200
tempX = np.linspace(5000,25000,21)
figures = QTable({'temp':tempX, 'theta':5040/tempX})
figures['boltz'] = 4*np.exp(-(-13.6/4 + 13.6)/(8.617* 10** -5 * figures['temp']))
```

```python
fig, ax = plt.subplots(
    figsize = (15, 8),
    constrained_layout = True
)

ax.set_title('N2/(N1+N2) vs Temperature')
ax.set_xlabel('Temperature in K')
ax.set_ylabel('N2/(N1+N2) of H')
ax.scatter(figures['temp'], figures['boltz']);
```

```python
saha = sahaEqn(xion, figures['temp'], figures['theta'], np.log10(PE))
figures['n2n1'] = 10**saha
#print(saha)
```

```python
figures['n2ntot'] = figures['n2n1']/(1 + figures['n2n1'])
```

```python
fig, ax = plt.subplots(
    figsize = (15,8),
    constrained_layout = True
)

ax.set_xlabel('Temperature in K')
ax.set_ylabel('N2/Ntot of H')
ax.set_title('N2/Ntot vs Temperature')
ax.scatter(figures['temp'], figures['n2ntot']);
```

```python
figures['n1ntot'] = 1/figures['n2n1']/(1 + 1/figures['n2n1'])
figures['n2ntotBoltz'] = figures['boltz'] * figures['n1ntot']
```

```python
fig, ax = plt.subplots(
    figsize = (15,8),
    constrained_layout = True
)

ax.set_xlabel('Temperature in K')
ax.set_ylabel('N2/Ntot of H')
ax.set_title('N2/Ntot vs Temperature')
ax.scatter(figures['temp'], figures['n2ntotBoltz']);
```
