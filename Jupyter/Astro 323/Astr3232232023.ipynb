```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z,tz,tk,da,dl=np.loadtxt('distances_NWCC.txt',delimiter=',',unpack=True)
```

```python
plt.plot(z,tk)
plt.grid();
```

```python
plt.plot(tz,(1/(1+z)))
plt.plot(tz,(tz/16.1)**(2/3))
plt.grid();
```

```python
plt.plot(np.log10(z),np.log10(da))
plt.plot(np.log10(z),np.log10(da*(1+z)**2))
plt.grid();
```

```python
plt.plot(np.log10(z),np.log10(dl))
plt.plot(np.log10(z),np.log10(da))
plt.grid();
```

Objects appear to be smaller the farther away they are which makes sense, however at a certain point objects that are very far away they appear to be closer because the light we measure was emitted when they were much closer to us.
